promedio ponderado

```python
egresados_compu = calculo_15_años_computacion['egresados_en_15_anios_o_menos'].sum()
egresados_biologia = calculo_15_años_biologia['egresados_en_15_anios_o_menos'].sum()
egresados_fisica = calculo_15_años_fisica['egresados_en_15_anios_o_menos'].sum()

calculo_15_años_computacion['tasa_egreso_15_años_con_d'].mean() * egresados_compu/total_egresados + calculo_15_años_biologia['tasa_egreso_15_años_con_d'].mean() * egresados_biologia/total_egresados + calculo_15_años_fisica['tasa_egreso_15_años_con_d'].mean() * egresados_fisica/total_egresados
```
